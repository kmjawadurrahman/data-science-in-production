### Load the Boston Housing dataset

The dataset has been removed from sklearn for ethical reasons. We shouldo only use it for educational purposes.
Got the dataset from Kaggle. Link: https://www.kaggle.com/code/prasadperera/the-boston-housing-dataset/data

Added the dataset to my Github repo and parsed the csv to ingest as pandas dataframe.

```python
data_url = "https://raw.githubusercontent.com/kmjawadurrahman/data-science-in-production/main/data/housing.csv.xls"
raw_df = pd.read_csv(data_url, sep="\s+", header=None)
```

```python
boston_df = raw_df.iloc[:, :-1]
boston_df.columns = load_boston().feature_names
```

```python
target_df = pd.DataFrame(raw_df.iloc[:, -1])
target_df.columns = ['label']
boston_df = pd.concat([boston_df, target_df], axis=1)
```
